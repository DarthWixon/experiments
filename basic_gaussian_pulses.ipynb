```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gauss(x, mean, std_dev):
    coeff = 1/(std_dev * np.sqrt(np.pi * 2))
    term = np.exp(-0.5 * ((x-mean)/std_dev)**2)
    return coeff * term
```

```python
a = np.linspace(-10, 10, 100)
```

```python
b = gauss(a, 0, 1)
```

```python
plt.plot(b)
plt.show()
```

```python
c = gauss(a, 3, 5)
```

```python
plt.plot(b)
plt.plot(c)
plt.show()
```
